```python
import pandas as pd
import os

# 파일 리스트
file_names = ['겐코.csv', '디왈리.csv', '육연차.csv', '치미치미.csv','푸드타운.csv', '호식당.csv','화로상회.csv', '화리화리.csv']

# 데이터프레임을 저장할 리스트
dataframes = []

# 각 파일을 읽어서 데이터프레임 리스트에 추가
for file in file_names:
    if os.path.exists(file):  # 파일이 존재하는지 확인
        df = pd.read_csv(file, encoding='utf-8-sig')  # CSV 파일 읽기
        df['Restaurant'] = os.path.splitext(file)[0]  # 파일명에서 확장자를 제거하고 식당 이름 추가
        dataframes.append(df)
    else:
        print(f"파일을 찾을 수 없습니다: {file}")

# 모든 데이터프레임을 결합
combined_df = pd.concat(dataframes, ignore_index=True)

# 결합된 데이터 확인
print(combined_df.head())

# 결합된 데이터 저장
output_file = 'combined_reviews.csv'
combined_df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"결합된 데이터가 저장되었습니다: {output_file}")
```
